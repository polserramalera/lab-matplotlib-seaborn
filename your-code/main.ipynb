#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here
fit_months = fitbit.groupby("Months_encoded")["Steps"].median()
fit_months

```

```python
# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here
fit_weekend = fitbit.groupby("Work_or_Weekend")["Steps"].median()
fit_weekend
fitbit[fitbit["Work_or_Weekend"] == 0].groupby(fitbit['Days'])['Steps'].count()

fitbit[fitbit["Work_or_Weekend"] == 0].groupby(fitbit['Days'])['Steps'].count()
fitbit["Steps"].groupby(fitbit["Work_or_Weekend"]).median()
```

```python
plt.style.use('classic')
Fig3,axes3=plt.subplots(1,2,figsize=(12,4))
Fig3.suptitle("Challenge 2",fontsize=18)

axes3[0].plot(fit_months.index,fit_months,color="green",label="Month")
axes3[0].set_title("Months")
plt.sca(axes3[0])
plt.xticks(list(range(13)), ["","Jan","Feb","Mar","Apr","May","Jun","Jul","Aug","Sep","Oct","Nov","Dec"], rotation = 'horizontal')


axes3[1].bar(fit_weekend.index,fit_weekend,color="blue",label="Weekend")
axes3[1].set_title("Weekend")
plt.sca(axes3[1])
plt.xticks(range(2), ["Weekend","Workdays"], rotation = 'horizontal')
plt.show
```
